```python
list1 = []
neg_count = 0
num = 0

number = int(input ("Please enter total number of elements: "))
for i in range(1, number +1):
    value = int(input("Please enter the value of %d element : " %i))
    list1.append(value)

for j in range(number):
    if (list1[j] < 0):
             neg_count += 1
            
    num += 1
      

print("Negative numbers in the list: ", neg_count)
```
